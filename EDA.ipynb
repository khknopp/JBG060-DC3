```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_dir = "data/" # Define data path
pf = pd.read_csv(data_dir + "articles_topics.csv") # Read data into DataFrame
df = pf.copy()
df["date"] = pd.to_datetime(df["date"]) # Create date column
print(len(df))
```

```python
# Add a new column 'year_month' with the year and month extracted from 'date'
df['year_month'] = df['date'].dt.strftime('%Y-%m')
```

```python
# Group by 'year_month' and sum the True values for each topic
grouped = df.groupby('year_month')[['hunger', 'refugees', 'humanitarian', 'conflict']].sum()

# Plot the results
grouped.plot(kind='line', marker='o')
plt.title('Number of Rows with True Values Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Count')
plt.legend(title='Topics', loc='upper right')
plt.grid(True)


plt.show()
```

```python
# Count the number of True values in the 'hunger' column
num_true_hunger = df['hunger'].sum()

# Count the number of True values in the 'refugees' column
num_true_refugees = df['refugees'].sum()

# Count the number of True values in the 'humanitarian' column
num_true_humanitarian = df['humanitarian'].sum()

# Count the number of True values in the 'conflict' column
num_true_conflict = df['conflict'].sum()

# Print the results
print(f"Number of rows with True values in 'hunger': {num_true_hunger}")
print(f"Number of rows with True values in 'refugees': {num_true_refugees}")
print(f"Number of rows with True values in 'humanitarian': {num_true_humanitarian}")
print(f"Number of rows with True values in 'conflict': {num_true_conflict}")

all_topicized = num_true_conflict+num_true_refugees+num_true_humanitarian+num_true_conflict
```

```python
print(f"Out of 18520 articles only {all_topicized} where assigned to one of the 4 topics")
```
